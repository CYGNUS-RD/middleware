```python
import pandas as pd

filename = "df_condor_coda3"

# Step 1: Read the JSON file using pandas
json_file_path = filename + '.json'  # Replace 'input.json' with the path to your JSON file
df = pd.read_json(json_file_path, orient="table")


```

```python
# Step 2: Save the DataFrame as a CSV file
csv_file_path = filename + '.csv'  # Replace 'output.csv' with the desired CSV file path
df.to_csv(csv_file_path, index=False)  # Set index=False to exclude the DataFrame index in the CSV file

print(f'JSON file "{json_file_path}" has been converted and saved as CSV file "{csv_file_path}"')
```
